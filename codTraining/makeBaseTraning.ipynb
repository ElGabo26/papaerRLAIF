```python
import pandas  as  pd
import os
from itertools import product, chain
```

```python
def makeBase(x, idres, idgrades, pares):
    res=[x[i] for i in idres]
    grades=[x[i] for i in idgrades]
    data=dict(zip(grades,res))
    result=[(x[0],data[p[0]],data[p[1]]) for p in pares]
    return result
```

```python
basesRoots=[f"./calify/{x}" for x in os.listdir("./calify")]
bases=[pd.read_csv(x) for x in basesRoots]
```

```python
responses=[f'response_{x}' for x in range(1,5)]
califications=[f'calificacion_response_{x}' for x in range(1,5)]
cols=list(bases[0].columns)
responsesId=[cols.index(x) for  x in responses]
calificationsId=[cols.index(x) for  x in califications]
#pares ordenados  que indican la  mejor  calificacion  1  es la mejor  4  es la peor
pares=list(product(range(1,5),range(1,5)))
pares=[x for x in pares if  x[0]<x[1]]
print(pares)

#creamos la  base  para  casa modelo
depseek=bases[0]
llama=bases[1]
qwen=bases[2]

columnsR=['prompt','chosen','rejected']
resultDeep=list(chain.from_iterable(map(
    lambda x: makeBase(x,responsesId,calificationsId, pares),
    depseek.values
)))
resultDeep=pd.DataFrame(columns=columnsR, data=resultDeep)
print(resultDeep.shape)
resultDeep=resultDeep[resultDeep['chosen']!=resultDeep['rejected']]
print(resultDeep.shape)

resultLlama=list(chain.from_iterable(map(
    lambda x: makeBase(x,responsesId,calificationsId, pares),
    llama.values
)))
resultLlama=pd.DataFrame(columns=columnsR, data=resultLlama)
print(resultLlama.shape)
resultLlama=resultLlama[resultLlama['chosen']!=resultLlama['rejected']]
print(resultLlama.shape)

resultQwen=list(chain.from_iterable(map(
    lambda x: makeBase(x,responsesId,calificationsId, pares),
    qwen.values
)))
resultQwen=pd.DataFrame(columns=columnsR, data=resultQwen)
print(resultQwen.shape)
resultQwen=resultQwen[resultQwen['chosen']!=resultQwen['rejected']]
print(resultQwen.shape)
# depseekPrefer=
```

```python
#guardamos los   resultados

resultDeep.to_csv("./calify/preferencesDeep.csv")
resultLlama.to_csv("./calify/preferencesLlama.csv")
resultQwen.to_csv("./calify/preferencesQwen.csv")
```
