```python
import pandas as pd
import os
```

```python
db=pd.read_csv(r"C:\Users\Gabo\Desktop\PAPPER Rlaif\createDataset\baseCabezales.csv", index_col=0)
```

```python
db.columns
```

```python
level=db[['pregunta','level']].copy()
level_ot=level[level['level']=='OTRO'].sample(6500)
level_0=level[level['level']!='OTRO'].copy()
level_0.replace({
    'MIDDLE':"OTRO",
    'HIGH':'OTRO'
}, inplace=True)
level_1=pd.concat([level_0, level_ot])
level_1.rename(columns={
    'pregunta':'text',
    'level':'labels'
}, inplace=True)


skill=db[['pregunta','skill']].copy()
skill_0=skill[skill['skill']!='otro']
skill_ot=skill[skill['skill']=='otro'].copy().sample(1100)
skill_1=pd.concat([skill_0,skill_ot])
skill_1.rename(columns={
    'pregunta':'text',
    'skill':'labels'
}, inplace=True)


claridad=db[['pregunta','claridad']].copy()
claridad.rename(columns={
    'pregunta':'text',
    'claridad':'labels'
}, inplace=True)

subject=db[['pregunta','subject']].copy()
subject_1=subject.copy()
subject_1['subject']=subject_1['subject'].apply(
    lambda x: 'otro' if x !='mathematics' else x
)
subject_1.rename(columns={
    'pregunta':'text',
    'subject':'labels'
}, inplace=True)
```
